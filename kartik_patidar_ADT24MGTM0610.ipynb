```python
plt.figure(figsize=(10,5))
sns.countplot(x='Attrition',data=df)
```

```python
attrition_counts = df['Attrition'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['red', 'yellow'])
plt.title('Attrition Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Department',data=df)
```

```python
Department_counts = df['Department'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(Department_counts, labels=Department_counts.index, autopct='%1.1f%%', startangle=90, colors=['red', 'yellow','blue'])
plt.title('Department Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='JobRole',data=df)
```

```python
JobRole_counts = df['JobRole'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(JobRole_counts, labels=JobRole_counts.index, autopct='%1.1f%%', startangle=90, colors=['red', 'yellow','blue'])
plt.title('JobRole Distribution')
plt.show()
```

```python
attrition_by_department = df.groupby('Department')['Attrition'].value_counts(normalize=True).unstack()
attrition_by_department['Attrition_Rate'] = attrition_by_department['Yes'] * 100
print("Attrition Rate by Department:")
display(attrition_by_department[['Attrition_Rate']].sort_values(by='Attrition_Rate', ascending=False))
```

```python
plt.figure(figsize=(10, 6))
attrition_by_department['Attrition_Rate'].plot(kind='bar')
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.show()
```
